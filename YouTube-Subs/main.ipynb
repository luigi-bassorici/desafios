```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Nova era
```

```python
data = pd.read_csv('data/topSubscribed.csv')
```

```python
# data
data.info()
```

## Número de Inscritos

```python
# transformando o número de inscritos em 'int64'
data.loc[:, "Subscribers"] = data.loc[:, "Subscribers"].str.replace(',', '').astype(int)
```

```python
# mostrando a distribuição de frequências
Subscribers = data.loc[:, "Subscribers"]
print(Subscribers.value_counts().head(10))
plt.subplots(figsize=(16,4))
plt.hist(data.loc[:, "Subscribers"], bins=40)
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.title('Histograma de Valores Inteiros')
plt.show()
```

```python
# Distribuição dos dados | separatrizes | outliers
Q1 = np.percentile(Subscribers, 25)
Q2 = np.percentile(Subscribers, 50)
Q3 = np.percentile(Subscribers, 75)
FIQ = Q3 - Q1

# Definindo os limites de outliers
lower_bound = Q1 - 1.5 * FIQ
upper_bound = Q3 + 1.5 * FIQ

# Exibindo os resultados
print(f"Quartil 1: {Q1}")
print(f"Quartil 2: {Q2}")
print(f"Quartil 3: {Q3}")
print(f"IQR: {FIQ}")
print(f"Limite inferior de outliers: {lower_bound}")
print(f"Limite superior de outliers: {upper_bound}")
plt.figure(figsize=(20, 2))
sns.boxplot(data.loc[:, "Subscribers"], orient='h')
sns.despine()
plt.xlabel('Subscribers')
plt.ylabel('Valor')
plt.title('Box-Plot dos Valores na Série Subscribers')
plt.show()
```
